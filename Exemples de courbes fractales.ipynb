# 1. Le flocon de Koch

```python
import turtle as tl
```

```python
def koch(l,n):
    if n==0:
        tl.forward(l)
    else:
        koch(l/3, n-1)
        tl.left(60)
        koch(l/3, n-1)
        tl.right(120)
        koch(l/3, n-1)
        tl.left(60)
        koch(l/3, n-1)
    return None
```

```python
def flocon(l,n):
    tl.reset()
    tl.speed(0)
    tl.up()
    tl.goto(-l/2, l/3)
    tl.down()
    tl.color("red")
    koch(l, n)
    tl.right(120)
    koch(l, n)
    tl.right(120)
    koch(l, n)
    tl.ht()
    tl.done()
    return None
```

```python
flocon(300,5)
```

# 2. Le triangle de Sierpinski

```python
def serp(l,n):
    if n==1:
        tl.begin_fill()
        tl.forward(l)
        tl.right(120)
        tl.forward(l)
        tl.right(120)
        tl.forward(l)
        tl.right(120)
        tl.end_fill()
    else:
        serp(l/2,n-1)
        tl.forward(l/2)
        serp(l/2,n-1)
        tl.right(120)
        tl.forward(l/2)
        tl.left(120)
        serp(l/2,n-1)
        tl.left(120)
        tl.forward(l/2)
        tl.right(120)
    return None
```

```python
def sierpinski(l,n):
    tl.reset()
    tl.speed(0)
    tl.color("red")
    tl.fillcolor("yellow")
    tl.up()
    tl.goto(-l/2, l/2)
    tl.down()
    serp(l,n)
    tl.ht()
    tl.done()
    return None
```

```python
sierpinski(450,7)
```

# 3. L'arbre de Pythagore

```python
def branche(l,n):
    if n==0:
        tl.color('green')
        tl.forward(l)
        tl.backward(l)
    else:
        tl.color('brown')
        tl.pensize(n)
        tl.forward(2*l/3)
        tl.left(30)
        branche(2*l/3,n-1)
        tl.right(60)
        branche(2*l/3,n-1)
        tl.left(30)
        tl.backward(2*l/3)
```

```python
def arbre(l,n):
    tl.reset()
    tl.speed(0)
    tl.penup()
    tl.goto(0,-l)
    tl.setheading(90)
    tl.pendown()
    branche(l,n)
    tl.ht()
    tl.done()
```

```python
arbre(200,8)
```

# 4. L'ensemble de Mandelbrot

```python
import matplotlib.pyplot as pl
from numpy import *
```

```python
def mandelbrot(n_iter):
    lignes = 240
    colonnes = 320
    matrice = zeros((lignes,colonnes))
    for x in range(colonnes):
        for y in range(lignes):
            z = complex(0,0)
            c = complex(3.5*x/colonnes-2.5,-2.5*y/lignes+1.25)
            i = 0
            while (i < n_iter) and (abs(z) < 2):
                i += 1
                z = z*z + c
            matrice[y,x] = i
    pl.figure(figsize=(10, 8))
    pl.axis("equal")
    pl.imshow(matrice)
    pl.show()
```

```python
mandelbrot(20)
```
